# 2: Quantum-Clustering

```python
from renom_q import *
from renom_q.ml.qclustering import QClustering
import numpy as np
```

 Perform Data clustering using adiabatic quantum computation by Ising model.

- The code of Quantum-Clustering  
 https://github.com/ReNom-dev-team/ReNomQ/blob/release/0.2/renom_q/ml/qclustering.py

Draw a graph with `QClustering.plot_graph(data)`.

```python
data_xy = np.array([[5, 4], [6, 6], [4, 9], [8, 1], [9, 7], [1, 4], [4, 6]])

qcl = QClustering()
qcl.plot_graph(data_xy)
```

Cluster data with `QClustering.qclustering(data, distance)`. Calculation is performed using the distance of each data point. When `distance = 'euclidean'` is input, clustering is performed by Euclidean distance, and when `distance = 'mahalanobis'` is input, clustering is performed by Mahalanobis distance.

### Example1: In the case of using the Euclidean distance

```python
label = qcl.qclustering(data_xy, 'euclidean')
print(label)
```

```python
qcl.plot_graph(data_xy, label)
```

### Example2: In the case of using the Mahalanobis distance

```python
label = qcl.qclustering(data_xy, 'mahalanobis')
print(label)
```

```python
qcl.plot_graph(data_xy, label)
```

By displaying `QClustering.result`, the calculation result of clustering can be viewed.

```python
print(qcl.result)
```

***

When drawing a quantum circuit, if you first call `QClustering` like `QClustering(set_qasm=True)`, qasm code is added to `QClustering.result`.
