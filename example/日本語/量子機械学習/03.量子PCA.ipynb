# 3: 量子主成分分析(PCA)

この量子主成分分析(PCA)のデモで用いた量子回路は、2019年度 人工知能学会全国大会（第33回）にて発表する[『量子コンピューティングにおける最適解の全量子探索』](https://confit.atlas.jp/guide/event/jsai2019/subject/2E1-J-1-02/advanced)の内容に基づいています。

```python
from renom_q import *
from renom_q.ml.qpca import QPCA
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from matplotlib import rc
```

量子変分アルゴリズム(VQE)を利用して、画像データなどの主成分分析を行う。

- 量子主成分分析のコード  
 https://github.com/ReNom-dev-team/ReNomQ/blob/release/0.2/renom_q/ml/qpca.py

```python
images = load_digits()
mnist_image = images.images[4]

mnist_image = mnist_image/255

plt.imshow(mnist_image)
plt.colorbar()
plt.tick_params(bottom=False, left=False)
plt.tick_params(labelbottom=False, labelleft=False)
plt.show()
```

`QPCA.fit(image, method)`でデータの主成分分析を行う。最適化手法を`method = 'Nelder-Mead'`とするとNelder-Mead法で、`method = 'QPSO'`とすると量子-粒子群最適化法(QPSO)で計算可能。 n_components は求めるcomponentの数であり、指定しない場合、データの次元数となる。

### 3-1: Nelder-Mead法の場合

`QPCA.fit`の引数`steps`は、Nelder-Mead法での最適解探索回数。指定しない場合、`steps=50`となる。

```python
n_components = 6
qpca = QPCA(n_components=n_components)
qpca.fit(mnist_image, method='Nelder-Mead', steps=10)
```

`QPCA.eigenvalue`で主成分の固有値、`QPCA.components_`で主成分の固有ベクトルを表示可能。

```python
print(qpca.eigenvalue)
print()
print(qpca.components_)
```

```python
num = n_components

size = int(np.log2(mnist_image.shape[0]))
eigenvector_image = []
eigenvector_sum_image = [np.zeros((2**size, 2**size)) for i in range(num)]

for i in range(num):
    vec = np.outer(qpca.components_[i], qpca.components_[i])
    array = np.dot(mnist_image, vec)
    eigenvector_image.append(array)
    for j in range(num-i):
        eigenvector_sum_image[num-1-j] += array
    
rc('text', usetex=True)
title = []
for i in range(num):
    title.append(r'$H \left| v_'+str(i)+r' \right\rangle \left\langle v_'+str(i)+r' \right|$')
title.append(r'$H \left| v_0 \right\rangle \left\langle v_0 \right|$')
for i in range(num-1):
    title.append(r'$\displaystyle \sum^'+str(i+1)+r'_{j=0} H \left| v_j \right\rangle \left\langle v_j \right|$')

fig = plt.figure(figsize=(13,13))

for i in range(2*num):
    plt.subplot(2, num, i+1)
    if i < num:
        plt.imshow(eigenvector_image[i])
    else:
        plt.imshow(eigenvector_sum_image[i-num])
    plt.title(title[i])
    plt.tick_params(bottom=False, left=False)
    plt.tick_params(labelbottom=False, labelleft=False)

fig.tight_layout()
plt.show()
```

### 3-2: QPSOの場合

`QPCA.fit`の引数`n_particle`は、QPSOでの粒子の数。指定しない場合、`n_particle=100`となる。引数`steps`は、QPSOでの最適解探索回数。指定しない場合、`steps=50`となる。

```python
n_components = 6
qpca = QPCA(n_components=n_components)
qpca.fit(mnist_image, method='QPSO', n_particle=80, steps=20)
```

```python
print(qpca.eigenvalue)
print()
print(qpca.components_)
```

```python
num = n_components

size = int(np.log2(mnist_image.shape[0]))
eigenvector_image = []
eigenvector_sum_image = [np.zeros((2**size, 2**size)) for i in range(num)]

for i in range(num):
    vec = np.outer(qpca.components_[i], qpca.components_[i])
    array = np.dot(mnist_image, vec)
    eigenvector_image.append(array)
    for j in range(num-i):
        eigenvector_sum_image[num-1-j] += array
    
rc('text', usetex=True)
title = []
for i in range(num):
    title.append(r'$H \left| v_'+str(i)+r' \right\rangle \left\langle v_'+str(i)+r' \right|$')
title.append(r'$H \left| v_0 \right\rangle \left\langle v_0 \right|$')
for i in range(num-1):
    title.append(r'$\displaystyle \sum^'+str(i+1)+r'_{j=0} H \left| v_j \right\rangle \left\langle v_j \right|$')

fig = plt.figure(figsize=(13,13))

for i in range(2*num):
    plt.subplot(2, num, i+1)
    if i < num:
        plt.imshow(eigenvector_image[i])
    else:
        plt.imshow(eigenvector_sum_image[i-num])
    plt.title(title[i])
    plt.tick_params(bottom=False, left=False)
    plt.tick_params(labelbottom=False, labelleft=False)

fig.tight_layout()
plt.show()
```

`QPCA.mean_`でデータの平均を、`QPCA.get_covariance()`でデータの共分散行列を表示可能。

```python
print(qpca.mean_)
print()
print(qpca.get_covariance())
```
